#Theory Questions
1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to decide whether there’s enough evidence in sample data to support or reject a claim (null hypothesis) about a population. It compares observed results to what would be expected by chance and makes a decision based on a significance level.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
- The **null hypothesis ($H_0$)** is the default assumption that there is **no effect, no difference, or no relationship** in the population.

- The **alternative hypothesis ($H_1$ or $H_a$)** is the competing claim that **there is an effect, a difference, or a relationship**.

-  **Difference:**

  * $H_0$ assumes *status quo* (nothing new is happening).
  * $H_1$ assumes *something new or different* is happening.

- **Example:**
A medicine company claims their pill lowers blood pressure by **0 mmHg** on average.

  * $H_0$: Mean change = 0 (no effect).
  * $H_1$: Mean change ≠ 0 (there is an effect).

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
- The **significance level** (denoted as $\alpha$) is the **threshold probability** we set before testing, which defines how much risk of making a **Type I error** (rejecting a true null hypothesis) we are willing to accept.

  Common choices are $\alpha = 0.05$ (5%), $0.01$ (1%), or $0.10$ (10%).

- **Role in hypothesis testing:**

  * We compare the **p-value** (probability of seeing our data if $H_0$ is true) to $\alpha$.
  * **If p-value ≤ α** → Reject $H_0$ → Evidence supports $H_1$.
  * **If p-value > α** → Fail to reject $H_0$ → Not enough evidence for $H_1$.

- **Example:**
  If $\alpha = 0.05$ and p-value = 0.03, we reject $H_0$ because the result is unlikely (only 3% chance) under the assumption that $H_0$ is true.

  Think of $\alpha$ as the **cutoff line** for deciding whether results are statistically significant.

4. What are Type I and Type II errors? Give examples of each.
- In hypothesis testing:

  ### **Type I Error** — *False Positive*

  * You reject $H_0$ when it is actually true.
  * Probability of this happening = **α** (significance level).
  * **Example:**
    A COVID test says a person has the virus (positive result), but they are actually healthy.

  ### **Type II Error** — *False Negative*

  * You fail to reject $H_0$ when it is actually false.
  * Probability of this happening = **β** (beta).
  * **Example:**
    A COVID test says a person is virus-free (negative result), but they actually have the virus.

5.What is the difference between a Z-test and a T-test? Explain when to use
each.
- The **Z-test** and **T-test** are both hypothesis tests for comparing means, but they differ mainly in **sample size** and **knowledge of population standard deviation (σ)**.

  ### **Z-test**

  * **Used when**:

    1. **Population standard deviation (σ) is known**.
    2. Sample size is **large** (n ≥ 30, Central Limit Theorem applies).
  * **Distribution**: Uses the **standard normal distribution (Z-distribution)**.
  * **Example**:
    Testing whether the average height of 1,000 students (σ known) is 170 cm.

  ### **T-test**

  * **Used when**:

    1. **Population standard deviation (σ) is unknown**.
    2. Sample size is **small** (n < 30).
  * **Distribution**: Uses the **t-distribution**, which has heavier tails than the normal distribution to account for extra uncertainty.
  * **Example**:
    Testing whether the average exam score of 15 students differs from 75, when σ is unknown.




```python
#6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

n=10
size=1000
p=0.5

data=np.random.binomial(n,p,size)

plt.hist(data, bins=np.arange(-0.5, n+0.5, 1), density=True, edgecolor='black')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu=50
sigma=0.5
alpha=0.05

n=len(sample_data)
sample_mean=np.mean(sample_data)

z_score=(sample_mean-mu)/(sigma/np.sqrt(n))
p_value=2*(1-norm.cdf(abs(z_score)))

if p_value<alpha:
  print('Reject null hypothesis')
else:
  print('Fail to reject null hypothesis')

```

```python
#8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
mean_true = 50
std_dev = 5
n = 100

data = np.random.normal(mean_true, std_dev, n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
standard_error = sample_std / np.sqrt(n)

confidence_level = 0.95
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
margin_of_error = t_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

plt.figure(figsize=(8,5))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Data with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
#   in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z = (data - mean) / std
    return z

data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5]

z_data = z_scores(np.array(data))

print("Z-scores:", np.round(z_data, 3))

plt.figure(figsize=(8,5))
plt.hist(z_data, bins=10, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
